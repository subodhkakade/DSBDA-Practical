```python
attribute = df[['Age','Chol','MaxHR']]
```

```python
target = df['RestECG']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(attribute,target,test_size=0.30,random_state=0)
```

```python
len(train_x)
```

```python
len(test_x)
```

```python
train_x
#Checking if the samples are randomly selected
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
lr.fit(train_x,train_y)
#Fit takes training data as arguments
```

```python
lr_y_pred = lr.predict(test_x)
```

```python
lr_acc = accuracy_score(test_y,lr_y_pred)
lr_acc
```
