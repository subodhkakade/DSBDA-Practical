```python
#Line Plot using Seaborn
sns.lineplot(x='Age', y='RestBP', data = df)
plt.title('RestBP Vs Age')
```

```python
#Displot using Seaborn
sns.displot(df['Chol']);
```

```python
#Histogram Plot using Seaborn
sns.histplot(df['ChestPain'])
```

```python
#Scatter Plot using Seaborn
sns.scatterplot(x = 'Age', y = 'RestBP', data = df, hue = 'Sex')
```

```python
#Bar Plot using Seaborn
sns.barplot(y = 'Chol', x = 'ChestPain', data = df)
```

```python
#Pair Plot using Seaborn
sns.pairplot(df, hue = 'Sex')
plt.figure(figsize=(12,12))
```

```python
#Scatter Plot using Matplot 
plt.scatter(df['Age'], df['RestBP'], color = 'red')
plt.title('RestBP as per age')
plt.xlabel('Age')
plt.ylabel('RestBP')


#can change marker = 'v' or '.'
```

```python
#Line Plot using Matplot
plt.plot(df['RestBP'], df['Sex'])
```

```python
#Initializing a new df with Sex value counts
sex_df = pd.DataFrame(df['Sex'].value_counts())
sex_df
```

```python
#PiePlot using Matplot
plt.pie(sex_df['Sex'],labels = sex_df.index)
```

```python
#Bar Plot using Matplot
plt.bar(sex_df.index,sex_df['Sex'])
```

```python
#Pie Plot of ChestPain
data = [df.ChestPain[df.ChestPain == "typical"].count(),df.ChestPain[df.ChestPain == "asymptomatic"].count(),
       df.ChestPain[df.ChestPain == "nonanginal"].count(),df.ChestPain[df.ChestPain == "nontypical"].count()]
labels = ["nonanginal","typical","nontypical","asymptomaic"]
plt.pie(data,labels=labels)
plt.show()
```
